```python
import pandas as pd
import plotly.express as px
```

```python
df=pd.read_csv('/content/cleanedaskamanager.csv')
```

```python
tmp=df.groupby(by='Industry')['Annual_salary'].median().nlargest(10).to_frame()
fig=px.bar(tmp,y='Annual_salary',color=tmp.index,text=tmp['Annual_salary'])
fig.update_layout(template="ggplot2", title="Top 10 Industries With Higest Median Annual Salary")

fig.show()
```

```python
tmp=df.groupby(by='Industry')['Annual_salary'].median().nsmallest(10).to_frame()
fig=px.bar(tmp,y='Annual_salary',color=tmp.index,
           color_discrete_sequence=px.colors.qualitative.G10,
           text=tmp['Annual_salary'])
fig.update_layout(
    title_text='Top 10 Industries With Lowest Median Annual Salary',
    xaxis_title='Industry',
    yaxis_title='Median Annual Salary',
    
)
fig.show()
```

```python

tmp=df.groupby(by='Age')['Annual_salary'].median().sort_values().to_frame()
fig=px.bar(tmp,y='Annual_salary',color=tmp.index,
color_discrete_sequence=px.colors.qualitative.Light24,
           text=tmp['Annual_salary'])
fig.update_layout(
    title_text=' Median Annual Salary by Age',
    xaxis_title='Age',
    yaxis_title='Median Annual Salary',
    template="seaborn"
)
fig.show()
```

```python
state_codes = {'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR',
               'California': 'CA', 'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE',
               'District of Columbia': 'DC', 'Florida': 'FL', 'Georgia': 'GA', 'Hawaii': 'HI',
               'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA', 'Kansas': 'KS',
               'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
               'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS',
               'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV',
               'New Hampshire': 'NH', 'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY',
               'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK',
               'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
               'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT',
               'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV', 'Wisconsin': 'WI',
               'Wyoming': 'WY'}
df['State_US']=df['State_US'].apply(lambda x: state_codes[x])               
```

```python
tmp=df.groupby(by='State_US')['Annual_salary'].median().sort_values().to_frame()
fig = px.choropleth(tmp, locations=tmp.index, locationmode='USA-states', color='Annual_salary',
                    scope='usa', color_continuous_scale='Viridis')


fig.update_layout(title='Median Annual Salary by State')
                  
fig.show()
```

```python
tmp=df.groupby(by='Work_experience_overall')['Annual_salary'].median().sort_values().to_frame()
fig=px.bar(tmp,y='Annual_salary',color=tmp.index,
color_discrete_sequence=px.colors.qualitative.Set3,
           text=tmp['Annual_salary'])
fig.update_layout(
    title_text='Median Annual Salary by Work Experience Overall',
    xaxis_title='Experience Overall',
    yaxis_title='Median Annual Salary',
    template="seaborn"
)
fig.show()

```

```python
tmp=df.groupby(by='Work_experience_field')['Annual_salary'].median().sort_values().to_frame()
fig=px.bar(tmp,y='Annual_salary',color=tmp.index,
color_discrete_sequence=px.colors.qualitative.Safe,
           text=tmp['Annual_salary'])
fig.update_layout(
    title_text='Median Annual Salary by Work Experience in Field',
    xaxis_title='Experience in Field',
    yaxis_title='Median Annual Salary',
    template="simple_white"
)
fig.show()
```

```python
tmp=df.groupby(by='Education')['Annual_salary'].median().sort_values()\
.to_frame()
fig=px.bar(tmp,y='Annual_salary',color=tmp.index,
color_discrete_sequence=px.colors.qualitative.Vivid,
           text=tmp['Annual_salary'])
fig.update_layout(
    title_text='Median Annual Salary by Education',
    xaxis_title='Education',
    yaxis_title='Median Annual Salary',
    template="plotly_white"
)
fig.show()
```

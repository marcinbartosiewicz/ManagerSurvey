```python

def group_similiar_names(names, similarity):
    grups = []
    for name in names:
        added = False
        for grup in grups:
            if fuzz.token_sort_ratio(name, grup[0]) >= similarity:
                grup.append(name)
                added = True
                break
        if not added:
            grups.append([name])
    return grups

```

```python
def checksimilarvaluesincolumn(column,similarity_value):
  df[column]=df[column].str.lower().str.strip()
  groups_of_silimiar_names=group_similiar_names(df[column].unique(),similarity_value)
  morethanonememberofgroup=[group for group in groups_of_silimiar_names 
  if len(group)>1]
  for index,group in enumerate(morethanonememberofgroup):
    print(f'{group} - {index}')
  print('Is every group ok?(Press y/n)')
  flag=input()
  if flag=='n':
     print('Which groups are wrong?(Type with delimiter between)')
     wronggroups=[int(value) for value in input().split(',')]
     morethanonememberofgroup=[group for 
     index,group in enumerate(morethanonememberofgroup) if index not in 
     wronggroups]
  for group in morethanonememberofgroup:
      change=group[0]
      for element in group[1:]:
        df.loc[df[df[column]==element].index,column]=change 



  
```

```python

indexes=[]
for ind in df.Industry.unique():
  tmp=df.query("Industry==@ind")
  Q1=tmp.Annual_salary.quantile(.25)
  Q3=tmp.Annual_salary.quantile(.75)
  IQR=Q3-Q1
  indexes.append(tmp.query("Annual_salary<@Q1-1.5*@IQR | Annual_salary>@Q3+1.5*@IQR").index)
flat_list = list(itertools.chain(*indexes))
df.drop(index=flat_list,inplace=True)

```
